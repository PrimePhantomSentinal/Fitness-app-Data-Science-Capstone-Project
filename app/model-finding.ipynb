```python
import pandas as pd
from itertools import combinations
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error

# Prevent truncation and wrapping
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)

feature_cols = [col for col in no_outlier.columns if col != 'Absenteeism time in hours']
target = final_df['Absenteeism time in hours']
results = []

# Choose number of features in the combination (e.g., 4)
for combo in combinations(feature_cols, 6):
    X_combo = final_df[list(combo)]
    y = target

    X_train, X_test, y_train, y_test = train_test_split(X_combo, y, test_size=0.2, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    results.append({
        'Features': ", ".join(combo),  # Or use "\n".join(combo) if you prefer line breaks
        'R² Score': round(r2, 4),
        'MAE': round(mae, 2)
    })

results_df = pd.DataFrame(results).sort_values(by='R² Score', ascending=False)
print(results_df.head(10))
```
